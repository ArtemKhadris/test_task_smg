```python
"""
Data Loading and Initial Setup Module.

This module handles the import of necessary libraries, defines file paths,
and loads the raw housing data for Madrid for further processing.
"""

import pandas as pd
from pathlib import Path

# Define file paths for raw and processed data
# Using Pathlib for cross-platform path handling
RAW_DATA_PATH = Path("../data/raw/houses_Madrid.csv")  # Path to raw dataset
PROCESSED_DATA_PATH = Path("../data/processed/processed.csv")  # Path for processed output

# Load the raw dataset into a pandas DataFrame
# This reads the CSV file from the specified path
df = pd.read_csv(RAW_DATA_PATH)

# Display the first few rows of the DataFrame
# This provides a quick overview of the data structure and content
df.head()
```

```python
"""
Data Exploration and Basic Analysis Module.

This module performs initial exploratory data analysis (EDA) to understand
the dataset structure, statistical properties, and data quality issues.
"""

# Display basic information about the DataFrame
# This shows data types, non-null counts, and memory usage
df.info()

# Generate comprehensive summary statistics for all columns
# include="all" ensures both numeric and categorical columns are included
# .T transposes the result for better readability
df.describe(include="all").T

# Calculate and display missing values count for each column
# This helps identify data completeness issues and potential cleaning needs
df.isna().sum()
```

```python
"""
Target Variable Distribution Analysis Module.

This module visualizes the distribution of the target variable (buy_price)
to understand its statistical properties, spread, and potential outliers.
"""

import seaborn as sns
import matplotlib.pyplot as plt

# Create histogram with kernel density estimate for the target variable
# bins=50 provides detailed distribution view across price ranges
# kde=True adds smoothed density curve for better pattern recognition
sns.histplot(df["buy_price"], bins=50, kde=True)

# Add title for clear chart identification
plt.title("Distribution of Housing Prices")

# Display the visualization
plt.show()
```

```python
# Check buy price by area column
sns.histplot(df["buy_price_by_area"], bins=50, kde=True)
plt.title("Distribution of Housing Prices")
plt.show()
```

```python
"""
Numeric Features Distribution Analysis Module.

This module identifies all numeric columns in the dataset and creates
histograms to visualize their distributions and statistical properties.
"""

import numpy as np

# Identify all numeric columns in the DataFrame
# select_dtypes(include=np.number) filters for all numeric data types (int, float)
# .columns extracts the column names for subsequent analysis
numeric_cols = df.select_dtypes(include=np.number).columns

# Create histograms for all numeric features
# figsize=(15, 15) sets large figure size for clear visualization of multiple subplots
# bins=30 provides sufficient granularity to see distribution patterns
df[numeric_cols].hist(figsize=(15, 15), bins=30)

# Add overall title for the multi-plot figure
# suptitle (super title) places title above all subplots
plt.suptitle("Numeric Feature Distributions")

# Display the grid of histograms
plt.show()
```

```python
"""
Outlier Detection and Variable Spread Analysis Module.

This module creates boxplots for key numeric variables to identify
outliers, understand data spread, and analyze variable distributions.
"""

# Iterate through specified key numeric columns for detailed analysis
# These columns are selected as they are typically important in housing data analysis
for col in ["buy_price", "sq_mt_built", "n_rooms", "n_bathrooms", "n_floors"]:
    
    # Check if column exists in DataFrame to avoid KeyError
    if col in df.columns:
        
        # Create new figure for each boxplot with consistent sizing
        plt.figure(figsize=(6, 4))  # Optimal size for clear visualization
        
        # Generate boxplot for the current column
        # Boxplot shows median, quartiles, and potential outliers
        sns.boxplot(x=df[col])
        
        # Add descriptive title with variable name for easy identification
        plt.title(f"Boxplot: {col}")
        
        # Display the individual boxplot
        plt.show()
```

```python
"""
Correlation Analysis Module.

This module calculates and visualizes the correlation matrix between
all numeric variables to identify relationships and multicollinearity.
"""

# Create a large figure to accommodate the heatmap visualization
plt.figure(figsize=(15, 6))  # Wide format to fit all variable labels clearly

# Calculate correlation matrix for all numeric columns
# corr() computes Pearson correlation coefficients between all pairs of numeric variables
corr = df[numeric_cols].corr()

# Create heatmap visualization of the correlation matrix
# annot=True: Displays correlation values on each cell for precise reading
# fmt=".2f": Formats correlation values to 2 decimal places for clarity
# cmap="coolwarm": Uses blue-to-red color scheme (cool = negative, warm = positive correlation)
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")

# Add descriptive title to the heatmap
plt.title("Correlation Heatmap")

# Display the correlation heatmap
plt.show()
```

```python
"""
Boolean to Numeric Conversion and Updated Correlation Analysis Module.

This module converts boolean columns to integer format (0/1) and
recalculates the correlation matrix to include these converted variables.
"""

# Create a copy of the DataFrame to avoid modifying the original data
# This ensures data integrity and allows for safe transformations
df = df.copy()

# Convert all boolean columns to integer type (0 for False, 1 for True)
# This enables inclusion of boolean variables in correlation analysis
for col in df.select_dtypes(include=["bool"]).columns:
    df[col] = df[col].astype(int)

# Recalculate numeric columns list to include newly converted integer columns
# Now includes both original numeric types and converted boolean columns
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns

# Create updated correlation heatmap including converted boolean variables
plt.figure(figsize=(15, 6))

# Calculate correlation matrix with all numeric variables including converted booleans
corr = df[numeric_cols].corr()

# Generate heatmap with correlation values displayed
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")

# Update title to reflect inclusion of converted boolean variables
plt.title("Correlation Heatmap (with booleans converted to 0/1)")

# Display the updated correlation heatmap
plt.show()
```

```python
# 1. Remove columns where all values are NaN
df = df.dropna(axis=1, how="all")

# 2. Remove rows where all values are NaN
df = df.dropna(axis=0, how="all")

# 3. Fill remaining NaN values
for col in df.columns:
    if df[col].isna().sum() > 0:
        if df[col].dtype in [np.float64, np.int64]:
            # Numeric -> fill with median
            df[col] = df[col].fillna(df[col].median())
        else:
            # Categorical -> fill with mode (most frequent value)
            df[col] = df[col].fillna(df[col].mode()[0])

# Check again
df.isna().sum()
```

```python
# Save cleaned dataset
df.to_csv(PROCESSED_DATA_PATH, index=False)
print(f"Processed dataset saved to {PROCESSED_DATA_PATH}")
```
